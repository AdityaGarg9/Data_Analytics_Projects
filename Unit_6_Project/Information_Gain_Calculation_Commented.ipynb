```python
import pandas as pd # Import pandas, the library that helps format our data as a dataframe
import numpy as np # Commonly used as a dependency by other libraries
from scipy.stats import entropy # This will help us calculate entropy and information gain for each feature

df = pd.read_csv('/Users/adityagarg/Documents/11th Grade/(H) Data Analytics/Unit_6_Project/training_data.csv') ## Import our data csv and read it in as a pandas dataframe

print(df.shape) # Check to make sure we have the correct number of rows and columns, shape gives output with format (rows, columns)
df.head() # Check the first five rows of data to make sure they look right/correspond to the input csv
```

```python
# We will want to write a loop later that goes through each feature and calculates a separate information gain for each of them
# Therefore, we define a list of features here.
features = ['Categorical_Sepal_Length_Feature1', 'Categorical_Petal_Length_Feature2', 'Categorical_Petal_Width_Feature3']

# Calculate overall entropy using log base 3
value_counts = df['Type_of_Flower_Target'].value_counts(normalize=True) # Creates a frequency distribution for each type of flowers using a proportion (normalized) and stores it in a variable
overall_entropy = entropy(value_counts, base=3) # Takes in a frequency distribution and a logbase to calculate overall entropy
print(f'Overall Entropy of Dataset: {overall_entropy}')
```

```python
# Manual information gain calculation using log base 3
def info_gain_base3(df, feature, target): # Defines a function to calculate information gain for each feature using logbase 3, taking in the dataframe (overall dataset), feature, and target
    
    # Weighted entropy for each feature value
    weighted_entropy = 0 # Initialize the weighted entropy at 0, then we can add to it
    for value in df[feature].unique(): # Runs this loop for each unique feature value
        subset = df[df[feature] == value]
        weight = len(subset) / len(df) # Calculates proportion of the value being iterated on out of the total
        subset_counts = subset[target].value_counts(normalize=True) # Creates a frequency distribution for each value within the feature being iterated on
        weighted_entropy += weight * entropy(subset_counts, base=3) # Adds the entropy of this feature value to the weighted entropy
    
    return overall_entropy - weighted_entropy # Subtracts weighted entropy from the overall dataset entropy to calculate information gain (relies on overall_entropy being defined in the outer scope)

for feature in features: # Will execute this loop for each feature in the previously defined list 'features'
    gain = info_gain_base3(df, feature, 'Type_of_Flower_Target') # Stores the calculated information gain (using the target and the current feature from the current iteration of the loop) in a variable
    print(f'Information Gain for {feature}: {gain}') # Prints the value for each iteration of the loop, so we/the user can see the outputs (information gain of each feature)
```
