```python
import pandas as pd # Import pandas, the library that helps format our data as a dataframe
from info_gain import info_gain # This library is used to calculate information gain, given a feature and a target (both columns in a pandas dataframe)

df = pd.read_csv('/Users/adityagarg/Documents/11th Grade/(H) Data Analytics/Unit_6_Project/training_data.csv') # Import our data csv and read it in as a pandas dataframe

print(df.shape) # Check to make sure we have the correct number of rows and columns, shape gives output with format (rows, columns)
df.head() # Check the first five rows of data to make sure they look right/correspond to the input csv
```

```python
# We will want to write a loop later that goes through each feature and calculates a separate information gain for each of them
# Therefore, we define a list of features here.
features = ['Categorical_Sepal_Length_Feature1', 'Categorical_Petal_Length_Feature2', 'Categorical_Petal_Width_Feature3']

for feature in features: # Will execute this loop for each feature in the previously defined list 'features'
        gain = info_gain.info_gain(df[feature], df['Type_of_Flower_Target']) # Stores the calculated information gain (using the target and the current feature from the current iteration of the loop) in a variable
        print(f'Information Gain for {feature}: {gain}') # Prints the value for each iteration of the loop, so we/the user can see the outputs (information gain of each feature)
```
