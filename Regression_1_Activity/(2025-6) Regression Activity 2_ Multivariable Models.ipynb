**Part Two: Graphing in 3D:**

We can create a 3D plot (e.g., $\mathbb{R}^{3}$) as follows:

```python
fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')
ax.grid()
```

To see how to create a scatterplot in $\mathbb{R}^{3}$, create some random coordinates:

```python
x = np.random.random(50)
y = np.random.random(50)
z = np.random.random(50)
```

We can now plot each pair of points using the following code:

```python
fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')
ax.grid()

ax.scatter(x, y, z, c = 'r', s = 50)
ax.set_title('3D Scatter Plot')
ax.set_xlabel('x-values')
ax.set_ylabel('y-values')
ax.set_zlabel('z-values')
```

To plot a surface in 3D, such as a plane, we need to create a list of $(x, y)$ pairs that cover the entire grid we're interested in. To do so, we'll use the numpy function [meshgrid](https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html).

For example, run the following code:


```python
x = np.linspace(-3, 3, 7)
y = np.linspace(-2, 2, 5)

print(x)
print(y)
```

```python
X, Y = np.meshgrid(x, y)

plt.plot(X, Y, 'ob')

plt.show()
```

We can then define our $z$ values using the $X$ and $Y$ created by meshgrid, and create our first 3D surface:

```python
Z = 2*X + 3*Y

fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')
ax.grid()
ax.plot_surface(X, Y, Z)

ax.set_title('3D Scatter Plot')
ax.set_xlabel('x-values')
ax.set_ylabel('y-values')
ax.set_zlabel('z-values')
```

This is a plot of the plane defined by:
\begin{align*}
z = 2x + 3y
\end{align*}
