# **Linear Regression in Python**

We'll use a new library to conveniently create linear regression models using Python.

This library will be particularly useful, as it will allow us to use multiple explanatory variables and/or create a model of higher degree (quadratic, cubic, quartic, etc.).

```python
from sklearn.linear_model import LinearRegression
```

You can find some documentation for LinearRegression here:

https://scikit-learn.org/0.15/modules/generated/sklearn.linear_model.LinearRegression.html

To create your model, you'll use the function "fit." This is a special kind of function, called a "method" is connected to a particular object (in this case, LinearRegression()).

```python
reg_model = LinearRegression().fit(x, y)
```

Once we've created that model, we can use the model to make predictions, using the "predict" function.

```python
y_predicted = reg_model.predict(x)
```

Based on the demo code in the "Graphing in Python" section above, fill in the below code.

```python
plt.plot(x, y, 'or') #Put a line of code here to plot the data using red, circular points
plt.plot(x, y_predicted, color='blue')#Put a line of code here to plot the predicted values (e.g., the linear model) as a solid blue line.
plt.show()
```

If we want to see the y-intercept of our model, we can do so as follows:

```python
print(reg_model.intercept_)
```

To see the coefficients of the model use "coef" rather than "intercept."

```python
print(reg_model.coef_)
```

Note that there should only be one value printed by the code above, as in a single-variable linear function, there is only one coefficient, the slope. If we created a higher-degree model or a multi-variable function, there would be multiple coefficients--we'll see this later.
